###Start Time

```python
data.Start_Time = pd.to_datetime(data.Start_Time)
```

```python
data.Start_Time[0]
```

```python
sns.histplot(data.Start_Time.dt.hour, stat='probability', bins=24)
```

- A high percentage of accidents occur between 14:00 to 18:00  (probably adults and students in a hurry to return home after work or school respectively).
- Next highest percentage is between 6:00 to 9:00.

```python
sns.histplot(data.Start_Time.dt.dayofweek, stat='probability', bins=7)
```

- A high percentage of accident occur on weekdays with an increasing trend from Monday to Friday. Then there is a serious decrease in the number of accidents from Saturday to Sunday.

Is the distribution of accidents by hour the same on weekends as on weekdays?

```python
sunday_start_time = data.Start_Time[data.Start_Time.dt.dayofweek == 6]
sns.histplot(sunday_start_time.dt.hour, stat='probability', bins=24)
```

```python
monday_start_time = data.Start_Time[data.Start_Time.dt.dayofweek == 0]
sns.histplot(monday_start_time.dt.hour, stat='probability', bins=24)
```

On Sunday, the peek occur between 0:00 and 3:00 in night and between 12:00 and 18:00 ,unlike weekdays where number of accident between 0:00 and 3:00 are less (This may be because of people returning home from late night parties and clubs and most of them might have comsumed alcohol).

```python
sns.histplot(sunday_start_time.dt.month, stat='probability', bins=12)
```

In this data the accidents of January 2016 are missing so we got a little less bar for January than my expectation.

Adverse weather conditions in winter lead to more collisions historically. The months of November, December, and January consistently show more accident then other months.

## Weather Condition

```python
data.Weather_Condition.unique()
```

So from this array it is clear that accident happened in many sort of weather condition. Let find the top occuring weather condition.

```python
data.Weather_Condition.value_counts()
```

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
weather_data = data['Weather_Condition'].value_counts().to_dict()
wc = WordCloud(width = 900, height = 800,
background_color ='white',
min_font_size = 10).generate_from_frequencies(weather_data)
plt.imshow(wc)
plt.axis('off')
plt.show()
```

The most frequently used word is “Fair”, followed by “Mostly Cloudy” and then “Partly Cloudy”. Now, the question arises that how did most of the recorded accidents occur under such weather conditions?

## Bump

It is a binary(True/False) column indicating the presence of a speed bump nearby the spot of the accident.

```python
data['Bump'].value_counts().plot(kind='pie')
```

- Speed control mechanisms are inadequate in the recorded accidents.

## Crossing

The column records the presence of a crossing nearby at the point of the accident.

```python
data['Crossing'].value_counts().plot(kind="pie")
```

- The accidents mostly happened away from a crossing. This brings to the query of proper signaling systems in these areas.

## Traffic Signal

A binary column indicating the presence of a traffic signal nearby the spot of the accident.

```python
data['Traffic_Signal'].value_counts().plot(kind='pie')
```

- Clearly, poor traffic management is a cause for the chunk of the accidents.

## Summary and Conclusion

Insights:
1. Top 5 cities with most accidents are Miami, Los Angeles, Orlando, Dallas, Houston respectively.
2. Ney York City is the most polulated city of US, but it was not even among the top 50 cities in terms of the numbers of accidents reported. 
3. Only 34 cities have number of accidents values greater than 10,000 out of 11,682 cities.
4. Less than 4.4% cities have more than 1,000 accidents.
5. Over 1100 cities have reported just one accident. (need to investigate)
6. Top 5 states with most accidents are California, Florida, Texas, Oregon, Virginia.

8. A high percentage of accident occur on weekdays with an increasing trend from Monday to Friday. Then there is a serious decrease in the number of accidents from Saturday to Sunday.
9. On Sunday, the peek occur between midnight and 3:00 am in morning unlike weedays. This may be because of people returning home from late night parties and clubs and most of them might have comsumed alcohol.
10. The year-wise analysis seems to show missing data about certain months. So nothing much can be conclusively said about the overall increasing trend in the latter half of the year.
11. The weather conditions seem to be “Fair”, “Mostly Cloudy”, “Partly Cloudy” and "clear" in most cases. This does not explain much why most accident happen in fair condition.
12. Most accidents happened away from the main crossing where there was inadequate or absence of traffic management infrastructures like traffic signaling systems and speed bumps.


In conclusion, it can be said that while the percentage of accidents is low for most cities, it is important to hammer the point that this analysis points to the following-

*The majority of such accidents seemed to have happened due to the absence of proper traffic management mechanisms like traffic signals and bumps- particularly in areas away from the main crossings.*
